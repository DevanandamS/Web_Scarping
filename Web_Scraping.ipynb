```python
from bs4 import BeautifulSoup
from requests import get
import pandas as pd
from time import sleep
from random import randint
from IPython.core.display import clear_output
headers = {"Accept-Language": "en-US, en;q=0.5"}
pages = [str(i) for i in range(51,1500,50)]
requests = 0
names = []
years = []
imdb_ratings = []
run_tym = []
votes = []
genres = []
for page in pages:
    response = get('https://www.imdb.com/search/title?title_type=feature&release_date=2017-01-01,2018-12-31&start='+page+'&ref_=adv_nxt',headers = headers)
    sleep(randint(8,15))
    requests += 1
    print('Request:{}; Frequency: {} requests/s'.format(requests, requests))
    clear_output(wait = True)
    if response.status_code != 200:
        warn('Request: {}; Status code: {}'.format(requests, response.status_code))
    if requests > 30:
        warn('Number of requests was greater than expected.')  
        break 
    html_soup = BeautifulSoup(response.text, 'html.parser')
    movie_containers = html_soup.find_all('div', class_ = 'lister-item-content')
    for container in movie_containers:
        if container.find('div', class_ = 'ratings-metascore') is not None:
            name = container.a.text
            names.append(name)
            year = container.h3.find('span', class_ = 'lister-item-year text-muted unbold').text
            years.append(year[-5:-1])
            imdb = float(container.strong.text)
            imdb_ratings.append(imdb)
            run_time=container.find('span', class_ = 'runtime').text
            run_tym.append(run_time)
            vote = container.find('span', attrs = {'name':'nv'})['data-value']
            votes.append(int(vote))
            gen=container.find('span', class_ = 'genre').text
            genres.append((gen[1:]))
test_df = pd.DataFrame({'Movie': names,
                       'Year': years,
                       'IMDB': imdb_ratings,
                       'Run_Time': run_tym,
                       'Votes': votes,
                       'Genre': genres})
print(test_df.info())
#test_df.head(100)
test_df.to_csv('Data2.csv')
```
